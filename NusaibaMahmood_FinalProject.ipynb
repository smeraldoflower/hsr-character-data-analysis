## **Data Cleaning**
- Remove numerical combat data columns which are not required by this project.
- Clean dirty data (unnecessary strings attached to data)
- Combine and Standardize character data from the multiple data sets.
- Add version column
- Add gender column
- Add missing character entries
- Add Remembrance Path
- Standardize character and path names. Remove under scores from data entries.
- Remove duplicates and handle special entries (Characters with multiple Paths and Types)
- Handle any missing values or entries.

```python
genders = char_genders.copy()

# Rename Element column to Combat Type
genders.rename(columns={'Element': 'Combat Type'}, inplace=True)

# Columns that need to be cleaned and words that need to be removed
clean_cols = ["Character", "Rarity", "Combat Type"]
remove_words = ["Image", "★", "Star Rail - "]

# Remove unnecessary words
for col in clean_cols:
    for word in remove_words:
        genders[col] = genders[col].str.replace(word, "", regex=True)

# Drop duplicate rows (if any)
genders = genders.drop_duplicates()

# Remove duplicate entries within cells
# For each cell in the column,
# Uses regex to handle concatenated duplicates like "LightningLightning" by separating them.
# Splits the text into words, and removes duplicates using dict.fromkeys(),
# Then rejoins the words into a single string.
for col in clean_cols:
    genders[col] = genders[col].apply(
        lambda x: " ".join(dict.fromkeys(re.sub(r"(\b[a-zA-Z]+)(?=\1)", r"\1 ", str(x)).split()))
    )

# Remove unnecessary spaces before and after entries
for col in clean_cols:
    genders[col] = genders[col].str.strip()

# Clean up inconsistencies
genders["Character"] = genders["Character"].replace({'March 7th (Imaginary)':'March 7th (The Hunt)',
                                                         'Topaz' : 'Topaz & Numby',
                                                         'Imbibitor Lunae':'Dan Heng • Imbibitor Lunae'})
# Set the Character column as the index
genders = genders.set_index('Character')

genders
```

```python
characters = characters_B.copy()

# Columns that need to be cleaned and words that need to be removed
clean_cols = ["Character", "Rarity", "Path", "Combat Type"]
remove_words = ["Character", "Star", "Icon", "Path", "Medium", "Type", "Small"]

# Remove unnecessary words
for col in clean_cols:
    for word in remove_words:
        characters[col] = characters[col].str.replace(word, "", regex=True)

# Drop duplicate rows (if any)
characters = characters.drop_duplicates()

# Remove duplicate entries within cells
# For each cell in the column,
# Splits the text into words, and removes duplicates using dict.fromkeys(),
# Then rejoins the words into a single string.
for col in clean_cols:
    characters[col] = characters[col].apply(
        lambda x: " ".join(dict.fromkeys(str(x).split()))
    )

# Remove unnecessary spaces before and after entries
for col in clean_cols:
    characters[col] = characters[col].str.strip()

characters
```

```python
# Outer Join Merge to combine the two data frames
chars = pd.merge(characters, genders, on=['Character', 'Rarity', 'Combat Type'], how='outer')
chars
```

```python
# Get rows with any NaN values
nan_rows = chars[chars.isna().any(axis=1)]
nan_rows
```

```python
# Update Missing Values

# Dictionaries for missing values
missing_paths = {'Aglaea': 'Remembrance', 'Fugue': 'Nihility', 'The Herta': 'Erudition'}
missing_vers = {'Aglaea': 3.0, 'Fugue': 2.7, 'The Herta': 3.0}

# Update the Path and Version columns based on the dictionaries
for character, path in missing_paths.items():
    chars.loc[chars['Character'] == character, 'Path'] = path

for character, path in missing_vers.items():
    chars.loc[chars['Character'] == character, 'Version'] = path

chars
```

```python
# Handle Trailblazer Special Case for Gender

# Get the rows with 'Trailblazer' in the Character column
trailblazer_rows = chars[chars['Character'].str.contains('Trailblazer', na=False)]

# Duplicate rows for Female and Male genders
female_rows = trailblazer_rows.copy()
female_rows['Gender'] = 'Female'

male_rows = trailblazer_rows.copy()
male_rows['Gender'] = 'Male'

# Combine the original DataFrame chars with the duplicated rows
chars = pd.concat([chars, female_rows, male_rows], ignore_index=True)

# Add Trailblazer (Remembrance) for both genders
new_rows = [
            {'Character': 'Trailblazer (Remembrance)', 'Rarity': 5, 'Path': 'Remembrance', 'Combat Type': 'Ice', 'Version': 3.0, 'Gender': 'Female'},
            {'Character': 'Trailblazer (Remembrance)', 'Rarity': 5, 'Path': 'Remembrance', 'Combat Type': 'Ice', 'Version': 3.0, 'Gender': 'Male'}
           ]

# Append the new rows to chars
chars = pd.concat([chars, pd.DataFrame(new_rows)], ignore_index=True)

# Remove NaN Trailblazer rows
chars = chars[~(chars['Character'].str.contains('Trailblazer', na=False) & chars['Gender'].isna())]

# Remove duplicate rows
chars = chars.drop_duplicates()

chars
```

```python
# Sort the Data Frame
sorted_chars = chars.sort_values(by=['Version', 'Character', 'Rarity', 'Gender'], ascending=[True, True, False, True])
sorted_chars.reset_index(drop=True, inplace=True)
```

```python
# Standardize names of Characters with multiple paths on the same rarity

# Female Trailblazer rows
sorted_chars.loc[
    (sorted_chars['Character'].str.contains('Trailblazer', na=False)) & (sorted_chars['Gender'] == 'Female'),
    'Character'
] = 'Trailblazer (Stelle)'

# Male Trailblazer rows
sorted_chars.loc[
    (sorted_chars['Character'].str.contains('Trailblazer', na=False)) & (sorted_chars['Gender'] == 'Male'),
    'Character'
] = 'Trailblazer (Caelus)'

# March 7th rows
sorted_chars.loc[(sorted_chars['Character'].str.contains('March 7th', na=False)),'Character'] = 'March 7th'

# Standardize all data types in the Rarity column to integer type.
sorted_chars['Rarity'] = sorted_chars['Rarity'].astype(int)

sorted_chars = sorted_chars.drop_duplicates()

sorted_chars.set_index("Version")
```

### How many characters were released/will release in each major version update (1.x, 2.x, 3.x)?

```python
# Group by Version and count the unique characters
chars_by_version = sorted_chars.groupby('Version')['Character'].nunique().reset_index(name='Character Count')

# Sort by Version
chars_by_version.sort_values('Version')

# Pivot the table so Versions become columns
pivoted_vers = chars_by_version.pivot_table(values='Character Count', columns='Version', aggfunc='sum')

# Fill NaN with 0 (if any version has no data) and display the result
pivoted_vers = pivoted_vers.fillna(0).astype(int)
pivoted_vers
```

```python
# Group by major versions 1.x, 2.x, 3.x
chars_by_major_version = chars_by_version.groupby(chars_by_version['Version'].astype(str).str.split('.').str[0] + '.x')['Character Count'].sum()

# Plot bar chart
plt.figure(figsize=(10, 5))
chars_by_major_version.plot(kind='bar', color=['#e83e3e', '#d6c146', '#1fb6d1'])

# Add labels and title
plt.xlabel('Major Version')
plt.ylabel('Character Count')
plt.title('Character Count by Major Version')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Display the chart
plt.tight_layout()
plt.show()
```

### How many characters are of 5-star and 4-star rarity respectively?

```python
# Group by Rarity and count unique characters
chars_by_rarity = sorted_chars.groupby('Rarity')['Character'].nunique().reset_index(name='Character Count')
pivoted_rares = chars_by_rarity.pivot_table(values='Character Count', columns='Rarity', aggfunc='sum')
pivoted_rares = pivoted_rares.fillna(0).astype(int)
pivoted_rares
```

```python
# Extract data for the pie chart
rarity_labels = pivoted_rares.columns           # Rarity: 4-star, 5-star
rarity_counts = pivoted_rares.values.flatten()  # Character Counts for each rarity

# Plot pie chart
plt.figure(figsize=(5, 5))
plt.pie(
    rarity_counts,
    labels=None,                   # Suppress labels
    autopct='%1.1f%%',             # Display percentages
    startangle=90,                 # Rotate the pie chart
    colors=['#9565DC', '#F2B145']  # Custom colors
)

# Add Legend
plt.legend(
    rarity_labels, 
    title="Rarity", 
    loc="center left", 
    bbox_to_anchor=(1, 0, 0.5, 1)
)

# Add title
plt.title('Character Distribution by Rarity')

# Display the chart
plt.tight_layout()
plt.show()
```
